```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import regex as re
```

```python
def tidy_df(df):
    newdf1 = df.drop(['Unnamed: 0'], axis=1)
    newdf2=  newdf1.set_index(['Sample','CompoundClass'])
    return newdf2
```

```python
#All of the names of things 
figure1B_UNIV = 'InputFigure1B_IntermediateStep' +'UNIV' +'.csv'
figure1B_ALL = 'InputFigure1B_IntermediateStep' +'ALL' +'.csv'
figure1B_IOS = 'InputFigure1B_IntermediateStep' +'IOS' +'.csv'
```

```python
#Input from elsewhere
all = pd.read_csv(write_dir + figure1B_ALL) 
univ= pd.read_csv(write_dir + figure1B_UNIV) 
ios = pd.read_csv(write_dir + figure1B_IOS) 
all
```

```python
tidy_all = tidy_df(all)
tidy_univ = tidy_df(univ)
tidy_ios = tidy_df(ios)
 

all_sub = tidy_all.add_suffix('_A')
univ_sub = tidy_univ.add_suffix('_U')
ios_sub = tidy_ios.add_suffix('_I')
```

```python
all_univ = all_sub.join(univ_sub)
all_univ_ios = all_univ.join(ios_sub)


unique_cols= all_univ_ios.T.drop_duplicates().T
unique_cols_clean= unique_cols.fillna(0)
unique_cols_clean.rename(columns={"Position_A": "Position", "Depth_A": "Depth", "SiteName_A": "SiteName", "TotalM_A": "TotalM", "CommonM_IOS_A": "TotalIOS", 
                                  "CommonM_universal_A": "TotalU"}, inplace=True)
unique_cols_clean= unique_cols_clean.fillna(0)
cleanly = unique_cols_clean.drop(columns=["Position_I", "Depth_I", "SiteName_I", "CommonM_universal_I","TotalM_I","CommonM_IOS_I"])
display(cleanly)
```
